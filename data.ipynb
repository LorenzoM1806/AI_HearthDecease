```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('train_heart.csv')
data
```

```python
# Plot histograms
plt.figure(figsize=(18, 5))

plt.subplot(1, 3, 1)
plt.hist(data['Age'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

plt.subplot(1, 3, 2)
plt.hist(data['RestingBP'], bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Resting Blood Pressure')
plt.ylabel('Frequency')
plt.title('Resting Blood Pressure Distribution')

plt.subplot(1, 3, 3)
plt.hist(data['Cholesterol'], bins=20, color='salmon', edgecolor='black')
plt.xlabel('Cholesterol Level')
plt.ylabel('Frequency')
plt.title('Cholesterol Level Distribution')

plt.tight_layout()
plt.show()
```

```python
# Create two subplots
plt.figure(figsize=(18, 5))

# Plot histogram for people without heart disease
plt.subplot(1, 2, 1)
sns.histplot(data=data[data['HeartDisease'] == 0], x='Age', bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution (No Heart Disease)')

# Plot histogram for people with heart disease
plt.subplot(1, 2, 2)
sns.histplot(data=data[data['HeartDisease'] == 1], x='Age', bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution (Heart Disease)')

plt.show()
```

```python
# Count the number of males and females
gender_counts = data['Sex'].value_counts()

# Create the pie chart with counts and percentages
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=[f'{gender} ({count})' for gender, count in gender_counts.items()], autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral'])
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Distribution of Men and Women')
plt.show()
```

```python
# Bar charts for categorical variables
plt.figure(figsize=(16, 10))

plt.subplot(3, 2, 1)
sns.countplot(x='Sex', data=data, hue='HeartDisease')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Sex Distribution')

plt.subplot(3, 2, 2)
sns.countplot(x='ChestPainType', data=data, hue='HeartDisease')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.title('Chest Pain Type Distribution')

plt.subplot(3, 2, 3)
sns.countplot(x='FastingBS', data=data, hue='HeartDisease')
plt.xlabel('Fasting Blood Sugar > 120 mg/dl')
plt.ylabel('Count')
plt.title('Fasting Blood Sugar Distribution')

plt.subplot(3, 2, 4)
sns.countplot(x='RestingECG', data=data, hue='HeartDisease')
plt.xlabel('Resting ECG')
plt.ylabel('Count')
plt.title('Resting ECG Distribution')

plt.subplot(3, 2, 5)
sns.countplot(x='ExerciseAngina', data=data, hue='HeartDisease')
plt.xlabel('Exercise Induced Angina')
plt.ylabel('Count')
plt.title('Exercise Induced Angina Distribution')

plt.subplot(3, 2, 6)
sns.countplot(x='ST_Slope', data=data, hue='HeartDisease')
plt.xlabel('ST Slope')
plt.ylabel('Count')
plt.title('ST Slope Distribution')

plt.tight_layout()
plt.show()

# Box plots for numerical variables
plt.figure(figsize=(16, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='HeartDisease', y='RestingBP', data=data)
plt.xlabel('Heart Disease')
plt.ylabel('Resting Blood Pressure')
plt.title('Resting Blood Pressure Distribution')

plt.subplot(1, 2, 2)
sns.boxplot(x='HeartDisease', y='Cholesterol', data=data)
plt.xlabel('Heart Disease')
plt.ylabel('Cholesterol Level')
plt.title('Cholesterol Level Distribution')

plt.tight_layout()
plt.show()
```

```python
# Bar charts for Heart Disease distribution based on Chest Pain Type
plt.figure(figsize=(12, 6))
sns.countplot(x='ChestPainType', data=data, hue='HeartDisease', palette='muted')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.title('Heart Disease Distribution based on Chest Pain Type')
plt.legend(title='Heart Disease', loc='upper right', labels=['No', 'Yes'])
plt.show()

# Box plots for Resting Blood Pressure and Cholesterol level based on Chest Pain Type
plt.figure(figsize=(16, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='ChestPainType', y='RestingBP', data=data)
plt.xlabel('Chest Pain Type')
plt.ylabel('Resting Blood Pressure')
plt.title('Resting Blood Pressure Distribution based on Chest Pain Type')

plt.subplot(1, 2, 2)
sns.boxplot(x='ChestPainType', y='Cholesterol', data=data)
plt.xlabel('Chest Pain Type')
plt.ylabel('Cholesterol Level')
plt.title('Cholesterol Level Distribution based on Chest Pain Type')

plt.tight_layout()
plt.show()
```
